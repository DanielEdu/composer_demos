```python
import pendulum
import dateutil.parser
import datetime
```

```python
tz = pendulum.timezone('America/Lima')
```

```python
ts = "2020-07-15T02:45:59.713374+00:00"
ts2 = "2020-07-14T02:40:00+00:00"
print('TS -> '+ ts2)
ts_p = dateutil.parser.parse(ts2)
print("ts_p  + 1 -> ", ts_p)
tss = tz.convert(ts_p).isoformat()
timestamp = tss
print('timestamp -> ' + timestamp)
```

```python
tss = "2020-07-15T02:45:59.713374+00:00"
ts_p = dateutil.parser.parse(tss)+ datetime.timedelta(days=1)
tss = tz.convert(ts_p).isoformat("T","seconds")
tss
```
